##**Trends**

Aggregate by Years and Months

```python
labels = num_labels
labels.drop("RTP")
mydict = {key:pd.Series.sum for key in labels}
mydict["RTP"] = pd.Series.mean
grouped = df.groupby(["Year","Month"],as_index=False).agg(mydict).sort_values(["Year","Month"],ascending=True)
x =   pd.to_datetime(grouped[['Year', 'Month']].assign(Day=1))
grouped.insert(0,"Date",x)

#Feature Engineering
#Creating OVERALL RTP as the total WON / BET in monthly period
grouped["Overall RTP"] = round((grouped.Won / grouped.Bet)*100,2)
#Creating commissions columns
grouped["Commissions"] = (grouped.Rake - grouped["Buy In Bonus"]) * 0.4

new_labels = num_labels.to_list()
new_labels.append("Overall RTP")
new_labels.append("Commissions")

grouped
```

Plotting the **Monthly Rake** from 2022

```python
grouped['Label'] = grouped['Year'].astype(str) + '-' + grouped['Month'].astype(str).str.zfill(2)

#Medium all time rake array
y = np.full(grouped["Label"].shape,grouped.Rake.mean())

plt.figure(figsize=(12, 6))

#Plotting Medium all time rake array
plt.plot(grouped['Label'],y,label="Meadium Rake all Time",linestyle ="--")

#Plotting Monthly Rake
plt.plot(grouped['Label'], grouped['Rake'], linestyle='-', label='Monthly Rake')

#Ticks and grid
years = grouped['Year'].unique()
ticks = []
for i in years:
  tick = grouped[grouped.Year == i].index[0]
  ticks.append(tick)

plt.xticks(
    ticks=ticks,
    labels=years,
    fontsize=12
)
plt.gca().set_xticks(range(len(grouped['Label'])), minor=True)
plt.gca().set_xticklabels(grouped['Month'], minor=True,fontsize=7)

plt.tick_params(axis='x', which='major', length=20, width=3)
plt.tick_params(axis='x', which='minor', length=5, width=1,)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Titles
plt.xlabel('Year and Month')
plt.ylabel('Rake')
plt.title('Monthly Rake')
plt.legend()
plt.tight_layout()
plt.show()

plt.show()
```

> It's clearly visible a **descending trend** beacause until May 2023 the monthly rake is always above its all time mean.


>From then it starts to decrease

Consequently, the platform generates more player winnings, reducing RAKE. Additionally, the significant **rise** in **BUY IN BONUS** further impacts final commissions.
```python
players = df[last_period].groupby("Username", as_index=False)["Rake"].sum().nsmallest(5, "Rake")["Username"].tolist()

colors = ["blue", "red", "green","orange","green"]
players_dict = {}

fig, axes = plt.subplots(3, len(players), figsize=(25, 25))  # 3 righe: bar chart, pie chart, bar chart Provider
axes = axes.flatten()

for n, (player, color) in enumerate(zip(players, colors)):
    # Filtrare e aggregare i dati
    players_dict[player] = df[(df.Username == player) & (last_period)]\
        .groupby(["Gioco", "Provider"], as_index=False).agg(mydict)\
        .sort_values(["Bet"], ascending=False).head(5)

    # Convertire in formato long per il barplot (prima riga)
    melted_df = pd.melt(players_dict[player], id_vars=["Gioco"], value_vars=["Buy In", "Rake"])

    # Bar chart (prima riga) - Buy In e Rake per Gioco
    ax_bar = axes[n]
    sns.barplot(data=melted_df, x="Gioco", y="value", hue="variable", ax=ax_bar)
    ax_bar.set_title(f"Games  {player}  lowest Rake  second period")
    ax_bar.set_xticklabels(ax_bar.get_xticklabels(), rotation=45)
    ax_bar.set_ylabel(f"{player}", fontsize=14)
    ax_bar.legend()

    # Pie chart Provider (seconda riga) - Distribuzione giochi per Provider
    ax_pie_provider = axes[n + len(players)]
    provider_counts = players_dict[player]["Provider"].value_counts()  # Conta i giochi per Provider

    if provider_counts.sum() > 0:
        ax_pie_provider.pie(provider_counts, labels=provider_counts.index, autopct="%1.1f%%", colors=plt.cm.Paired.colors)
    else:
        ax_pie_provider.text(0, 0, "No data", ha='center', va='center', fontsize=12)

    ax_pie_provider.set_title(f"Provider distribution {player}")

    # Bar chart Provider (terza riga) - Totale Buy In e Rake per Provider
    ax_bar_provider = axes[n + 2 * len(players)]
    provider_data = players_dict[player].groupby("Provider")[["Buy In", "Rake"]].sum().reset_index()

    if not provider_data.empty:
        melted_provider = pd.melt(provider_data, id_vars=["Provider"], value_vars=["Buy In", "Rake"])
        sns.barplot(data=melted_provider, x="Provider", y="value", hue="variable", ax=ax_bar_provider)
        ax_bar_provider.set_xticklabels(ax_bar_provider.get_xticklabels(), rotation=45)
    else:
        ax_bar_provider.text(0, 0, "No data", ha='center', va='center', fontsize=12)

    ax_bar_provider.set_title(f" Provider stats for {player} in second period")

# Sistemare la spaziatura tra i grafici
plt.tight_layout()
plt.subplots_adjust(hspace=0.6)  # Aumentare lo spazio tra le righe
plt.subplots_adjust(wspace=0.8)  # Aumentare lo spazio tra le righe

plt.show()
```
